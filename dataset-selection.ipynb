```python
import json
import numpy as np
```

```python
dataset = json.load(open('./data/q&a_full.json', 'r'))
```

```python
dataset
```

```python
sources = [item['src'] for item in dataset]
np.unique(sources, return_counts=True)
```

```python
ids = np.random.choice(np.arange(len(dataset)), 50)
```

```python
output = []
for id in ids:
    output.append(dataset[id])
len(output)
```

```python
sources = [item['src'] for item in output]
np.unique(sources, return_counts=True)
```

```python
with open('./data/q&a.json', 'w') as f:
    json.dump(output, f, indent=2)
```
